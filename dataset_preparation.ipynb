```python
from datasets import load_dataset

ds = load_dataset("alibayram/identity_finetune_magibu_q3")
```

```python
ds
```

```python
ds['turkish'][0]['train']
```

```python
# Convert the dataset to JSON format and save it to a file

import json


with open("datasets/turkish_dataset.json", "w", encoding="utf-8") as f:
    json.dump(ds['turkish'].to_dict(), f, ensure_ascii=False, indent=4)
```

```python
with open("datasets/turkish_dataset.json", "r", encoding="utf-8") as f:
    tr_new_json_data = json.load(f)

tr_new_json_data
```

```python
with open("datasets/english_dataset.json", "r", encoding="utf-8") as f:
    en_new_json_data = json.load(f)

en_new_json_data
```

```python
from datasets import DatasetDict, Dataset

new_ds = DatasetDict({
    "turkish": Dataset.from_dict(tr_new_json_data),
    "english": Dataset.from_dict(en_new_json_data)
})

new_ds
```

```python
new_ds.push_to_hub("alibayram/yeni_dataset")
```

```python
dr_gpt5 = load_dataset("alibayram/doktorsitesi-gpt5")
dr_gpt5
```

```python
text = ""

for row in dr_gpt5['train']:
    text += row['question_content'] + "\n" + row['question_answer'] + "\n----\n"

with open("datasets/bpe-dataset-corpus.txt", "w", encoding="utf-8") as f:
    f.write(text)
```
